```python
import os
import serpapi
from dotenv import load_dotenv
import csv
import pandas as pd

load_dotenv()

#unique serpAPI key
api_key = os.getenv("SERPAPI_KEY")

#shoe names
shoe = "Jordan 5 Retro Oreo"
#all variations of shoe name
shoe_search = "Jordan 5 Retro Oreo, Jordan 5 Retro Mid Oreo"
shoe_filename = "Jordan5RetroOreo.csv"

#search for google trend data
client = serpapi.Client(api_key = api_key)
search = client.search(
    engine = "google_trends",
    q = shoe_search,
    date = "2024-03-01 2024-04-12",
    api_key = api_key
)

#extracts relevant data from search
results = search["interest_over_time"]["timeline_data"]

#writes shoe name, date, and relative interest into csv file
with open(shoe_filename, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)

    #write the headers
    csv_writer.writerow(["Shoe", "Date", "Relative Interest (out of 100)"])

    #write the data
    for line in results:
        #summed_value should reflect the number of name variations
        summed_value = line["values"][0]["extracted_value"] + line["values"][1]["extracted_value"]
        csv_writer.writerow([shoe, line["date"], summed_value])

#saves csv file as a pandas data frame to be viewed
data = pd.read_csv(shoe_filename, index_col = 0)
```
