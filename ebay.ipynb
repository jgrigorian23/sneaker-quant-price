```python
from bs4 import BeautifulSoup
import csv
import pandas as pd

# open the html file
with open('/content/jordan 5 retro oreo for sale _ eBay.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# parse html file
soup = BeautifulSoup(html_content, 'html.parser')

# anything with the <li> tag in the html file is going to be a shoe and its associated information. We want to find all of these and collect them in a variable that we can loop over.
ebayShoes = soup.find_all('li', class_='s-item s-item__pl-on-bottom')

# now that we have the shoes, we want to loop over them and save each shoe's information in the csv file
with open('/content/output.csv', 'w', newline='', encoding='utf-8') as csvfile:

    # setting up the csv file
    fieldnames = ['Title','Sold Date', 'Condition', 'Brand', 'Ratings', 'Price', 'Bids', 'Shipping']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    # for each shoe that got saved, we want to record all the associated information with that shoe in the csv file
    for item in ebayShoes:
        sold_date_elem = item.find('span', class_='POSITIVE')
        sold_date = sold_date_elem.text.strip() if sold_date_elem else 'N/A'

        title_elem = item.find('span', role='heading')
        title = title_elem.text.strip() if title_elem else 'N/A'

        condition_elem = item.find('span', class_='SECONDARY_INFO')
        condition = condition_elem.text.strip() if condition_elem else 'N/A'

        brand_elem = item.find(class_= "s-item__subtitle")
        brand = brand_elem.text.strip() if brand_elem else 'N/A'

        ratings_elem = item.find('span', class_='s-item__reviews-count')
        ratings = ratings_elem.text.strip() if ratings_elem else 'N/A'

        price_elem = item.find('span', class_='s-item__price')
        price = price_elem.text.strip() if price_elem else 'N/A'

        bids_elem = item.find('span', class_='s-item__bids s-item__bidCount')
        bids = bids_elem.text.strip() if bids_elem else 'N/A'

        shipping_elem = item.find('span', class_='s-item__shipping s-item__logisticsCost')
        shipping = shipping_elem.text.strip() if shipping_elem else 'N/A'

        # write all the associated information for a shoe as a row in the csv file
        writer.writerow({
            'Title': title,
            'Sold Date': sold_date,
            'Condition': condition,
            'Brand': brand,
            'Ratings': ratings,
            'Price': price,
            'Bids': bids,
            'Shipping': shipping
        })

# save csv file as a panda data frame that we can look at below
data = pd.read_csv("output.csv", index_col=0)
data
```
